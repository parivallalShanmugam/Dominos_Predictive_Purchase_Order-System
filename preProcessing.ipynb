### Date column handle

```python
def date_taime(text):
    try:
        return pd.to_datetime(text, format="%d-%m-%Y")
    except ValueError:
        try:
            return pd.to_datetime(text, format="%d/%m/%Y")
        except ValueError:
            return pd.NaT
df_sales['order_date'] = df_sales['order_date'].apply(date_taime)
```

### feature engineering for date column

```python
df_sales['year'] = df_sales['order_date'].dt.year.astype(int)
df_sales['month'] = df_sales['order_date'].dt.month.astype(int)
df_sales['day'] = df_sales['order_date'].dt.day.astype(int)
```

```python
df_sales['start_of_week'] = df_sales['order_date'].dt.day_of_week
```

### Time column dtype changed and creating Hours column

```python
df_sales['order_time'] = pd.to_timedelta(df_sales['order_time'])

df_sales['hour'] = df_sales['order_time'].dt.components['hours']
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

df_grouped = df_sales.groupby('order_date')['quantity'].sum().reset_index()

df_grouped.set_index('order_date', inplace=True)

plt.figure(figsize=(14, 6))
plt.plot(df_grouped.index, df_grouped['quantity'], color='blue', marker='.', linestyle='-', linewidth=1)

ax = plt.gca()
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1)) 
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y')) 
plt.title('Pizza Orders Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Quantity Sold', fontsize=12)
plt.xticks(rotation=45) 
plt.grid(visible=True)  
plt.tight_layout()  

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by 'order_date' and sum the quantities
df_grouped = df_sales.groupby('day')['quantity'].sum().reset_index()

# Set the 'order_date' as the index
df_grouped.set_index('day', inplace=True)

plt.figure(figsize=(10, 6))
plt.plot(df_grouped.index, df_grouped['quantity'], color='blue', marker='.', linestyle='-', linewidth=1)

# Add titles and labels
plt.title('Pizza Orders Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Quantity Sold', fontsize=12)
plt.xticks(rotation=45) 
plt.grid(visible=True)  
plt.tight_layout()  

plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller

# ACF and PACF plots
plot_acf(df_sales['quantity'], lags=30)
plt.title('ACF Plot')
plt.show()

plot_pacf(df_sales['quantity'], lags=30)
plt.title('PACF Plot')
plt.show()

# ADF Test
adf_result = adfuller(df_sales['quantity'])
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
```
