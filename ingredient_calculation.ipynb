```python
import pandas as pd
```

```python
df_ing = pd.read_csv('Pizza_ingredients.csv')
df_ing.head()
```

```python
df_sale_for = pd.read_csv('pizza_sales_forecast.csv')
df_sale_for.head()
```

```python
# Merge sales forecast with ingredients table on 'pizza_name_id'
df_total_ingredients = df_sale_for.merge(df_ing, on='pizza_name_id', how='left')

# Calculate total ingredients needed
df_total_ingredients['Total_Qty_Required'] = df_total_ingredients['yhat'] * df_total_ingredients['Items_Qty_In_Grams']

# Group by ingredients to get the total amount required for each
df_ingredient_summary = df_total_ingredients.groupby('pizza_ingredients')['Total_Qty_Required'].sum().reset_index()

# Display final ingredient requirements
df_ingredient_summary
```
